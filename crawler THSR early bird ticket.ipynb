```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#高鐵 (加不加headers都只能拿到這個資料)
url = "https://www.thsrc.com.tw/TimeTable/Search"
d = {
"SearchType": "S",
"Lang": "TW",
"StartStation": "TaiPei",
"EndStation": "TaiNan",
"OutWardSearchDate": "2020/12/22",
"OutWardSearchTime": "21:00",
"ReturnSearchDate": "2020/12/16",
"ReturnSearchTime": "22:00",
"DiscountType":""
}

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
          }

r = requests.post(url, data = d, headers = headers)
soup = bs(r.text,'html.parser')
tabs = soup.find("div", class_ = "tr-td")
```

```python
# import json
# rjson = r.json()
# rjson
rjson = r.json()
rjson
```

```python
row_list = []
for i in range(20):
    Train_num = rjson['data']['DepartureTable']['TrainItem'][i]['TrainNumber']
    Dep_time = rjson['data']['DepartureTable']['TrainItem'][i]['DepartureTime']
    Des_time = rjson['data']['DepartureTable']['TrainItem'][i]['DestinationTime']
    
    dis_list = []
    if len(rjson['data']['DepartureTable']['TrainItem'][i]['Discount']) == 0:
        Bird = "不打折"
    for j in range(len(rjson['data']['DepartureTable']['TrainItem'][i]['Discount'])):
        dis_name = rjson['data']['DepartureTable']['TrainItem'][i]['Discount'][j]['Name']
        
        if dis_name == "早鳥":
            Bird = rjson['data']['DepartureTable']['TrainItem'][i]['Discount'][j]['Value']
            break
        else:
            if j == len(rjson['data']['DepartureTable']['TrainItem'][i]['Discount'])-1:
                Bird = "不打折"
            pass
        
    dict1 = {'Train_num':Train_num, 'Dep_time': Dep_time, 'Des_time':Des_time, 'Bird' : Bird}
    row_list.append(dict1)
    
    #大學生折價太少了所以用早鳥票
```

```python
df = pd.DataFrame(row_list)
df
```
