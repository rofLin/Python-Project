```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red

#RGBCMYK
#Red
#Green
#Blue
#Cyan
#Magenta
#Yellow
#blacK
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
#axis is different from axes
```

```python
plt.plot(x, np.sin(x), "-g", label = 'sin(x)')
plt.plot(x, np.cos(x), ":c", label = 'cos(x)')
plt.axis('equal')

plt.legend()
```

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel("x")
plt.ylabel('sin(x)')
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim = (0,10), ylim = (-2,2),
      xlabel = 'x', ylabel = 'sin(x)',
      title = 'A Simple Plot')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
import numpy as np
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
plt.plot(x, y, '-ok'); #line(-) circle marker(o), black(k)
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
plt.scatter(x, y, marker = 'o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=color, s=sizes, alpha=0.3,
           cmap = 'viridis')  #0-1之間原本是灰階，colormap映到彩色上
plt.colorbar();
```

```python
from sklearn.datasets import load_iris


iris = load_iris()
features = iris.data.T #4x150

plt.scatter(features[0], features[1], alpha=0.3,
           s = 100*features[2], c = iris.target, cmap = 'viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
